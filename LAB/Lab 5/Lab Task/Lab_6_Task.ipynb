# Import necessary libraries

```python
# do
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact, interact_manual
%matplotlib inline
from numpy.fft import fft, fftshift, fftfreq, ifft, ifftshift
pi = np.pi
```

```python
def perform_fft(x, N):
    X = fft(x)
    X = fftshift(X)
    X = X/N # Normalization

    fk = fftfreq(N)
    fk = fftshift(fk)

    return X, fk

def plot_sig_and_fft(x, n, X, fk, f):
    fig = plt.figure(figsize=(10, 10))

    plt.subplot(1, 2, 1)
    plt.stem(n, x)
    plt.xlabel("$n$")
    plt.ylabel("$x[n]$")
    plt.title(f"sinusoid with f = {f}")

    plt.subplot(1, 2, 2)
    plt.stem(fk, np.abs(X))
    plt.xlabel("Digital frequncy, f")
    plt.title("Magnitude of X [normalized]")
```

# Task 1: 1D Fourier Transform

Define a function $2\cos(2\pi f_1n) + 3\cos(2\pi f_2n)$

```python
# Here,
# f1 = 0.3 cycles per sample
# f2 = 0.2 cycles per sample
# n = 0 to N-1
# N = 100




```

```python
# perform the fourier transform of the signal
# plot the signal and it's fourier magnitude spectrum side by side

```

```python
# in the code cell for Task 1, update your signal definition as follows:
n_cycles_1 = 30
n_cycles_2 = 20
N = 100

f1 = n_cycles_1 / N
f2 = n_cycles_2 / N
n = np.arange(N)
x = 2*np.cos(2*np.pi*f1*n) + 3*np.cos(2*np.pi*f2*n)  # adjusted coefficients

X, fk = perform_fft(x, N)
plot_sig_and_fft(x, n, X, fk, [f1, f2])
```

# Task 2: IFFT

```python
# performt the inverse fourier transform of X
# plot the original signal and the recosntructed signal side by side
```

```python
def perform_ifft(X, N):
    X_ishift = ifftshift(X)
    x_recon = ifft(X_ishift)
    x_recon = np.real(x_recon)*N

    n = np.arange(N)
    if x_recon.shape[0] > N:
        x_recon = x_recon[:N]

    return x_recon, n
```

```python
x_reconm, n_recon = perform_ifft(X, N)


fig = plt.figure(figsize=(16, 4))

plt.subplot(1, 2, 1)
plt.plot(n, x)
plt.xlabel("$n$")
plt.ylabel("$x[n]$")
plt.title(f"original")

plt.subplot(1, 2, 2)
plt.plot(n_recon, x_reconm)
plt.xlabel("$n$")
plt.ylabel("$x_{\mathrm{recon}}[n]$")
plt.title("Reconstructed")
plt.show()
```
